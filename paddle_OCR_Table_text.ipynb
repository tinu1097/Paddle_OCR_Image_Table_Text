```python
dij = 0
def text_im(image_path):
  global dij
  result = ocr.ocr(image_path,cls=True)
  s = []
  for idx in range(len(result)):
    res = result[idx]
    for line in res:
      # print(line[1])
      result_string = " ".join(str(elem) for elem in line[1][:-1])
      s.append(result_string)
      print(result_string)
  # Open the file in the specified mode
  with open('./Text_Files/Text_files'+"_"+str(dij), "a") as file:
    for i in s:
      file.write(i + "\n")
  dij+=1
  print("successfully saved file")

#funcrion for save the text files in the text_Files folder
for files in os.listdir("./Images_data"):
    if files.endswith(".png") or files.endswith(".jpg") or files.endswith(".jpeg"):
      text_im("./Images_data/" + files)
```

```python
di = 0
def table_im(image_path):
  global di
  img = Image(src=image_path)
  try:
    extracted_tables = img.extract_tables(ocr=paddle_ocr,
                                          implicit_rows=True,
                                          borderless_tables=True,
                                          min_confidence=20)
    df = extracted_tables[0].df
    file_path = f'./CSV_Files/CSV_file_{di}.csv'
    print("file_path",file_path)
    df.to_csv(file_path, index=False)
    print(f"CSV file saved successfully: {file_path}")
    di+=1
    print("di",di)
  except IndexError:
      print("Error: Index is out of range. Please provide a valid index.")
  except Exception as e:
    print(f"Error occurred: {str(e)}")
```

```python
for files in os.listdir("./Images_data"):
    if files.endswith(".png") or files.endswith(".jpg") or files.endswith(".jpeg"):
      table_im("./Images_data/" + files)
```
